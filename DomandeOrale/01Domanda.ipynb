```python
"""Domanda 1. Dato un problema test di dimensioni variabili 
                                    Ax = b
la cui soluzione esatta è il vettore x̄ = (1, . . . , 1)T e b è il termine noto, 
ottenuto come Ax̄ calcolare:
    • il numero di condizionamento della matrice A
    • la soluzione x del sistema lineare Ax = b mediante la fattorizzazione LU con pivoting.
Considerare come matrice A la matrice quadrata, di numeri casuali generati mediante la 
funzione randn di Numpy, con n variabile tra 10 e 1000."""
import numpy as np
import scipy.linalg.decomp_lu as LUdec 
from scipy.linalg import lu, solve_triangular
```

```python
n = np.random.randint(10, 1001)
A = np.random.randn(n, n)
x_exact = np.ones(n)
b = np.dot(A, x_exact)
print('n:', n, '\n')
print('Matrice A: \n', A, '\n')
print('x: \n', x_exact, '\n')
print('b: \n', b, '\n')

cond1 = np.linalg.cond(A, 1)
cond2 = np.linalg.cond(A, 2)
condfro = np.linalg.cond(A, 'fro')
condinf = np.linalg.cond(A, np.inf)

print ('K(A)_1 = ', cond1, '\n')
print ('K(A)_2 = ', cond2, '\n')
print ('K(A)_fro = ', condfro, '\n')
print ('K(A)_inf = ', condinf, '\n')

##LU pivoting

P, L, U = lu(A)
print("Matrice P (matrice di permutazione):")
print('\n', P, '\n')
print("Matrice L (matrice triangolare inferiore):")
print('\n', L, '\n')
print("Matrice U (matrice triangolare superiore):")
print('\n', U, '\n')
y = solve_triangular(L, np.dot(P, b), lower=True)
x = solve_triangular(U, y)

print("Soluzione calcolata:\n")
print(x)
```
