```python
"""Domanda 5 (Approssimazione minimimi quadrati). Discutere l’approssimazione a m dati 
equispaziati ottenuti campionando la funzione f (x) = exp(x/2) nell’intervallo [−1, 1] 
usando polinomi di grado n ∈ {1, 2, 3, 5, 7}.
Si risolva il problema dei minimi quadrati sia mediante le equazioni normali che con SVD, 
calcolando l’errore di approssimazione in norma 2 per entrambi i metodi. 
Infine graficare le approssimazioni polinomiali ottenute."""
import numpy as np
from scipy.linalg import cholesky
from numpy.linalg import svd, solve
import matplotlib.pyplot as plt
from numpy.linalg import norm
```

```python
def f(x):
    return np.exp(x/2)

a, b = -1, 1
m = 10
m_plot = 100
n_values = [1, 2, 3, 5, 7]

x_observed = np.linspace(a, b, m)
y_observed = f(x_observed)

def create_matrix_A(x, n):
    A = np.zeros((len(x), n+1))
    for i in range(n+1):
        A[:, i] = x**i
    return A

for n in n_values:
    print(f'\n data n: {n}\n')
    A = create_matrix_A(x_observed, n)
    x_exact = np.ones((n+1, 1))
    b = np.dot(A, x_exact)
    print(f'Matrice A:\n{A}\n')
    print(f'x: \n{x_exact}\n')
    print(f'b: \n{b}\n')

    print(f'Metodo dei minimi quadrati con Cholesky per n={n}:\n')
    L = cholesky(np.dot(A.T, A), lower=True)
    x_chol = solve(L.T, solve(L, np.dot(A.T, b)))
    alpha_chol = solve(L, x_chol)
    error_chol = norm(np.dot(A, alpha_chol) - y_observed, 2)
    print(f'Soluzione alpha calcolata (Cholesky): \n{alpha_chol}')
    print(f'Errore approssimazione metodo Cholesky: {error_chol}\n')

    print(f'Metodo dei minimi quadrati con SVD per n={n}:\n')
    U, s, Vh = svd(A)
    alpha_svd = np.zeros(n+1)
    for i in range(n+1):
        ui = U[:, i]
        vi = Vh[i, :]
        alpha_svd = alpha_svd + (np.dot(ui, y_observed) * vi) / s[i]
    error_svd = norm(np.dot(A, alpha_svd) - y_observed, 2)
    print(f'Soluzione alpha con SVD: {alpha_svd}\n')
    print(f'Errore approssimazione metodo SVD: {error_svd}\n')

    x_plot = np.linspace(x_observed[0], x_observed[-1], m_plot)
    A_plot = np.zeros((m_plot, n+1))
    for i in range(n+1):
        A_plot[:, i] = x_plot**i 
    y_interpolation = A_plot @ alpha_svd 
    
    plt.plot(x_observed, y_observed, 'o') 
    plt.plot(x_plot, y_interpolation, 'r') 
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Interpolazione polinomiale di grado {n}')
    plt.grid()
    plt.show()
```
